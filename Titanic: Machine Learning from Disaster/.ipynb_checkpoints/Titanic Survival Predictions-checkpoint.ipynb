```python
import numpy as np
import pandas as pd

from sklearn.tree import DecisionTreeClassifier

pd.options.mode.chained_assignment = None  # disable the chain assignment warning
```

```python
# Load the train and test datasets into two DataFrames
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
# view the dataset
train.head()
```

```python
# Description of the dataset
train.describe()
```

```python
# Passengers' survival
print('Passenger:')
print(train['Survived'].value_counts(normalize=True))

# Males' survival
print('\nMale:')
print(train['Survived'][train['Sex'] == 'male'].value_counts(normalize=True))

# Females' survival
print('\nFemale:')
print(train['Survived'][train['Sex'] == 'female'].value_counts(normalize=True))
```

```python
# Check for children

# create column
train['Child'] = np.NaN

# assign values
# 1) child: age < 18
# 2) adult: age >= 18
train['Child'][train['Age'] < 18] = 1
train['Child'][train['Age'] >= 18] = 0

# Child survival
print('Children:')
print(train['Survived'][train['Child'] == 1].value_counts(normalize=True))

# Adult survival
print('\nAdult:')
print(train['Survived'][train['Child'] == 0].value_counts(normalize=True))
```

```python
train.head()
```

```python
# Cleaning and formatting the training data

# Convert gender to binary
# Male: 0
# Female: 1
train['Sex'][train['Sex'] == 'male'] = 0
train['Sex'][train['Sex'] == 'female'] = 1

# Fill the NaN values of Emabrked with most common value
train['Embarked'] = train['Embarked'].fillna(max(train['Embarked'].value_counts().index))

# Convert the Embarked classes to integer form
train["Embarked"][train["Embarked"] == "S"] = 0
train["Embarked"][train["Embarked"] == "C"] = 1
train["Embarked"][train["Embarked"] == "Q"] = 2

# Fill the NaN values of Age and Child with median value
train['Age'] = train['Age'].fillna(train['Age'].median())
train['Child'] = train['Child'].fillna(train['Child'].median())
```

```python
# Extracting useful features and creating a target variable
target = train['Survived'].values
features_one = train[['Pclass', 'Sex', 'Age', 'Fare']].values
```

```python
# Fit the first decision tree
my_tree_one = DecisionTreeClassifier()
my_tree_one = my_tree_one.fit(features_one, target)
```

```python
# Importance and the score of the included features
print('Feature Importance: ', my_tree_one.feature_importances_)
print('Score:', my_tree_one.score(features_one, target))
```

```python
# View the test data
test.head()
```

```python
# Describe the test set
test.describe()
```

```python
# Filling up NaN entries in test data

# Impute the missing values with the median
test.Fare[152] = test['Fare'].median()
test['Age'] = test['Age'].fillna(test['Age'].median())

# Convert gender to binary
# Male: 0
# Female: 1
test['Sex'][test['Sex'] == 'male'] = 0
test['Sex'][test['Sex'] == 'female'] = 1
```

```python
# Extract the useful features from the test set
test_features = test[['Pclass', 'Sex', 'Age', 'Fare']].values
```

```python
# Predict the survival rate
my_prediction = my_tree_one.predict(test_features)
```

```python
# Create a data frame with two columns: PassengerId & Survived.
# Survived contains the predictions
PassengerId = np.array(test['PassengerId']).astype(int)
my_solution = pd.DataFrame(my_prediction, PassengerId, columns = ['Survived'])
```
